# Pre-processing

```python
texts= filter_text['Text'].values.tolist()
texts
```

```python

def clean_tweet(tweet):
    '''
    Function to clean a tweet text by removing the links, punctuations, hashtags, and mentions symbols
    using regex expressions.
    Parameters:
       tweet (string): Tweet text to transform.
    Returns:
       tweet_cleaned (string): Tweet without links, punctuations, hashtags, and mentions symbols.
    """
    '''
    # convert the tweet text to lowercase letters
    #tweet_lower = tweet.lower()

    #remove mentions
    remove_mentions = re.sub("@[A-Za-z0-9_]+","", tweet)

    #remove hashtag
    remove_hastag = re.sub("#[A-Za-z0-9_]+","", remove_mentions)

    # #remove non ascii
    # remove_others = re.sub("[^\x00-\x7F]+", "", remove_mentions)

    # remove links, punctuations, hashtags, and etc
    text = re.sub(u"(http\S+)|([‚Äú‚Äù?])|([#|_*@()])|", "", remove_hastag)

    # remove \n
    text = re.sub(r" & ", " and ", text)
    text = re.sub(r"&", " ", text)
    text = re.sub(r"[\/]", " ", text)
    text = re.sub(r"=", " equal ", text)
    text = re.sub(r"\n", "", text)
    text = re.sub(r":", " ", text)
    text= re.sub(" [ ]+", " ", text)

    return text
```

```python
tweets_text_cleaned = [clean_tweet(text) for text in texts]
print(tweets_text_cleaned[:10])
```

```python
cleaned_tweets = pd.DataFrame({'Cleaned_Text': tweets_text_cleaned})
cleaned_tweets
```

# Vader

## Sentiment between emotes

```python
tweets_vader = filter_text
tweets_vader_cleaned = cleaned_tweets
```

```python
sid = SentimentIntensityAnalyzer()

def sentiment_analyzer_scores(sentence):
    score = sid.polarity_scores(sentence)
    print("{:-<40} {}".format(sentence, str(score)))

print(sentiment_analyzer_scores('üíï '))
print(sentiment_analyzer_scores('üòÅ '))
print(sentiment_analyzer_scores('üéâ'))
print(sentiment_analyzer_scores('üíî'))
print(sentiment_analyzer_scores('üò¨'))
print(sentiment_analyzer_scores('üôÅ'))
```

```python
tweets_vader['vader'] = tweets_vader['Text'].apply(lambda x: sid.polarity_scores(x))
# Extract the polarity score from the dictionary
tweets_vader['compound'] = tweets_vader['vader'].apply(lambda score_dict: score_dict['compound'])
tweets_vader.head()
```

```python
tweets_vader['Vader_Sentiment_Type']=''
tweets_vader.loc[tweets_vader.compound>0,'Vader_Sentiment_Type']='POSITIVE'
tweets_vader.loc[tweets_vader.compound==0,'Vader_Sentiment_Type']='NEUTRAL'
tweets_vader.loc[tweets_vader.compound<0,'Vader_Sentiment_Type']='NEGATIVE'
tweets_vader.head()
```

```python
tweets_vader_cleaned['vader'] = tweets_vader_cleaned['Cleaned_Text'].apply(lambda x: sid.polarity_scores(x))
# Extract the polarity score from the dictionary
tweets_vader_cleaned['compound'] = tweets_vader_cleaned['vader'].apply(lambda score_dict: score_dict['compound'])
tweets_vader_cleaned.head()
```

```python
tweets_vader_cleaned['Vader_Sentiment_Type']=''
tweets_vader_cleaned.loc[tweets_vader_cleaned.compound>0,'Vader_Sentiment_Type']='POSITIVE'
tweets_vader_cleaned.loc[tweets_vader_cleaned.compound==0,'Vader_Sentiment_Type']='NEUTRAL'
tweets_vader_cleaned.loc[tweets_vader_cleaned.compound<0,'Vader_Sentiment_Type']='NEGATIVE'
tweets_vader_cleaned.head()
```

```python
df1 = tweets_vader_cleaned.drop(['Cleaned_Text'], axis=1)
df2 = tweets.drop(['Unnamed: 0', 'Text'], axis=1)
```

```python
tweets_vader = pd.merge(df2, df1, left_index=True, right_index=True)

tweets_vader.rename({'english-Only-Cleaned_Text': 'English_Only_Cleaned_Text'}, axis=1, inplace=True)
tweets_vader
```
